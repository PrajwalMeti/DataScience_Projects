```python
# 2. Check for correlation of each input variable with the output(HasDetections).
def drop_the_columns():
    output = []
    [ output.append(col) for col in microsoft_df.columns if len(microsoft_df[col].unique()) == microsoft_df.shape[0]]  
    [ output.append(col) for col in microsoft_df.columns if len(microsoft_df[col].unique()) == 1]
    return output
drop_the_columns() 
microsoft_df = microsoft_df.drop(columns = ['MachineIdentifier', 'IsBeta', 'AutoSampleOptIn'])
```

```python
def corelated_or_not(columns):
    output = {}
    x = [col for col in columns]
    observed_table = pd.crosstab(microsoft_df[col],microsoft_df['HasDetections'])
    chi2_contingency(observed_table)
    p_value =chi2_contingency(observed_table)[1]
    if p_value >0.05 : 
        output = {microsoft_df[col]:'Yes'}
    else:
        output = {microsoft_df[col] :'No'}
    return output
corelated_or_not(microsoft_df.columns)
```
