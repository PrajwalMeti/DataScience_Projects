```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
from scipy.stats import chi2_contingency
from scipy.stats import f_oneway
```

```python
microsoft_df = pd.read_csv('Microsoft_data.csv')
```

```python
unique_variables = ([col for col in microsoft_df.columns if len(microsoft_df[col].value_counts()) == microsoft_df.shape[0]])
```

```python
same_variables =  ([col for col in microsoft_df.columns if len(microsoft_df[col].value_counts()) == 1])
same_variables
```

```python
microsoft_df.drop(columns= same_variables + unique_variables , inplace= True)
```

```python
columns_non_object = list([ col for col in microsoft_df.columns if microsoft_df[col].dtype != 'object'])
```

```python
microsoft_df[columns_non_object] = microsoft_df[columns_non_object].astype('object')
```
