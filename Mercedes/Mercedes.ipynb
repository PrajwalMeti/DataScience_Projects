```python
import pandas as pd 
import numpy as np 
```

```python
mercedes_df = pd.read_csv('Mercedes.csv')
```

```python
# missing value percentage in each columns
empty_columns = mercedes_df.isnull().sum()
empty_columns
```

```python
def columns_wise_null_percentage(col):
    output = []
    for column in col:
        output.append((column/mercedes_df.shape[0])*100)
    return zip(mercedes_df.columns,output)
empty_col_percentage = list(columns_wise_null_percentage(empty_columns))  
empty_col_percentage
```

```python
def find_the_unique(columns):
    return list(zip(mercedes_df.columns, [mercedes_df[i].unique() for i in columns if (len(mercedes_df[i].unique())) == mercedes_df.shape[0]  ]))

find_the_unique(mercedes_df.columns)
```

```python
mercedes_df.drop(columns = 'ID',inplace = True)
```

```python
def find_the_same(col):
    output = []
    for i in col:
        if len(mercedes_df[i].unique()) == 1:
            output.append(i)
    mercedes_df.drop(columns = output ,inplace = True)
x = find_the_same(mercedes_df.columns)
print(x)
```

```python
"""
    1. Identify & Drop all the columns with NA % > 50
	2. Identify & drop the columns which have ALL UNIQUE values or ALL SAME.
	3. Check for the correlation between each input & output variable(y).
	4. Check for the distribution of y- identify the outlier values(in a list).
    Check for skewness, if skewed- normalize the column.
    
"""
```

```python
from scipy.stats import f_oneway
columns = mercedes_df.columns
def anova(col):
    Correlated = []
    for columns in col:
        categories = list(mercedes_df[columns].value_counts().index)
        p_value = f_oneway(*[mercedes_df[mercedes_df[columns] == cat]['y'] for cat in categories])[1]

        Correlated.append(p_value)
    return Correlated
z = anova(columns)
print(list(zip(columns,z)))
```
