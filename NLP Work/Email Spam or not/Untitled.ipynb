```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
```

```python
email_df = pd.read_csv('spam.csv')
email_df['Category'].value_counts()
```

```python
def get_spam(x):
    if x == 'spam':
        return 1
    else:
        return 0 
```

```python
email_df['spam'] = email_df['Category'].apply(get_spam)
```

```python
email_df.drop(columns='Category',inplace = True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(email_df['Message'], email_df['spam'],test_size= 0.2)
```

```python
X_train.shape
```

```python
v = CountVectorizer()
v.fit_transform(X_train)
```

```python
X_train_cv = v.fit_transform(X_train.values)
X_test_cv = v.transform(X_test.values)
```

```python
v.get_feature_names_out().shape
```

```python
dir(v)
```

```python
v.vocabulary_
```

```python
mnb = MultinomialNB()
mnb.fit(X_train_cv,y_train)
```

```python
y_pred = mnb.predict(X_test_cv)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
```
