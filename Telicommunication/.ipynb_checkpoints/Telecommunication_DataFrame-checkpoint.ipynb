```python
def outliers(dataframe,column):
    q3 = dataframe[column].describe()['75%']
    q1 = dataframe[column].describe()['25%']
    IQR = q3-q1 
    HVO = q3 + (1.5 * IQR)
    LVO = q3 - (1.5 * IQR)
    outlier  = []
    for i in dataframe[column]:
        if i > HVO:
            outlier.append(i)
        if i < LVO:
            outlier.append(i)
    return(outlier)
outliers(telecom_df,'TotalCharges')
```

```python
# The output is Catogorical So we have to find the Co-relationship between 
#the Annova(Corelation betwween Continious and Catogorical Variables)

# Chi Squere Test for the Co-relation between the Catogorical Varaibles
cat_columns = cat_columns[:-1:]
for column in cat_columns:
    observed_table  = pd.crosstab(telecom_df[column],telecom_df['Churn'])
    if chi2_contingency(observed_table)[1] < 0.05:
        print(f"{column}: Corelatted")
    else:
        print(f"{column}: Not Corelated")   
```

```python
# Anova is for the Corelation between the Continous and Catagorical Variables So 
def Anova(telecom_df,cont_columns):
    for column in cont_columns:
        catogories = list(telecom_df['Churn'].value_counts().index)
        p_value = f_oneway(*[telecom_df[telecom_df['Churn'] == col][column] for col in catogories])[1]
        if p_value < 0.05:
            print(f"{column}: Corelated")
        else:
            print(f"{column}: Not Corelated")
Anova(telecom_df,cont_columns)
```
