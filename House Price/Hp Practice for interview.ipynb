```python
import pandas as pd 
import numpy as np 
from scipy.stats import chi2_contingency,f_oneway
```

```python
hp_df = pd.read_csv('house_price_data.csv')
hp_df
```

```python
observed_table = pd.crosstab(hp_df['MSZoning'],hp_df['Street'])
```

```python
pvalue = chi2_contingency(observed_table)[1]
if pvalue < 0.05:
    print('corelated')
else:
    print('not')
```
